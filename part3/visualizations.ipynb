```python
import pandas as pd
import numpy as np
import matplotlib

filepath = "data/CS170_Spring_2023_Small_data__95.txt"

df = pd.read_csv(filepath, delim_whitespace=True, header=None, engine="python")

features = [0, 4, 5]
subsetDf = df.iloc[:,features]
plot1 = subsetDf.plot.scatter(4, 5, c=subsetDf.iloc[:,0], colorbar=True)
plot1.set_xlabel("Feature 4")
plot1.set_ylabel("Feature 5")
plot1.set_title("Small Dataset 95 with features [4, 5]")
```

```python
filepath = "data/CS170_Spring_2023_Large_data__95.txt"
df = pd.read_csv(filepath, delim_whitespace=True, header=None, engine="python")

features = [0, 10, 19]
subsetDf = df.iloc[:,features]
plot2 =subsetDf.plot.scatter(10, 19, c=subsetDf.iloc[:,0], colorbar=True)
plot2.set_xlabel("Feature 10")
plot2.set_ylabel("Feature 19")

plot2.set_title("Large Dataset 95 with features [10, 19]")
```
